## Twitter API

```python
# Iqbal
CONSUMER_KEY = ''
CONSUMER_SECRET_KEY = ''

ACCESS_TOKEN = ''
ACCESS_SECRET_TOKEN = ''
```

```python
AUTH = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET_KEY)
AUTH.set_access_token(ACCESS_TOKEN, ACCESS_SECRET_TOKEN)

# Initialize API
API = tweepy.API(AUTH)
```

## Getting the Tweet

```python
QUERY = 'muslim OR muslim\'s OR islam OR islams' # Ganti query
TWEETS = 2000 # Tweets per run. [Kalo gasalah 900 tweets maximum per request.]
RUNS = 1 # Run yang mau dijalanin.
UNTIL_DATE = "2022-10-14" # Tanggal tweet terakhir data yang ke scrap adalah H-1 dari tanggal ini
```

```python
df_tweets = pd.DataFrame(columns = [ 'texts', 'created_at','location'])

p_start = time.time()
for i in range(0, RUNS):
  start_run = time.time()

  tweet_list = tweepy.Cursor(API.search_tweets, q=QUERY + ' -filter:retweets',
                              lang="id", until=UNTIL_DATE, tweet_mode='extended').items(TWEETS)
    
  num_tweets_added = 0
  for tweet in tweet_list:
    # print(json.dumps(tweet._json, indent=2))
    # break

    # Adding elements
    created = datetime.datetime.strftime(tweet.created_at, "%Y-%m-%d")
    # username = tweet.user.screen_name
    location = tweet.user.location
    # retweet =  tweet.retweet_count

    try:
      text = tweet.retweeted_status.full_text
    except:
      text = tweet.full_text
  
    n_tweets = [text, created, location]
    df_tweets.loc[len(df_tweets)] = n_tweets

    num_tweets_added += 1

  end_run = time.time()
  interval = round((end_run-start_run)/60, 2)

  print('RUN {}'.format(i+1))
  print('=====')
  print('Num Tweets Scraped {}'.format(num_tweets_added))
  print('Time {}'.format(interval))

  # if i+1 != RUNS:
  #   time.sleep(920) # Kalo lebih dari sekali run uncomment ini.
```

```python
def unique(list1):
  
    # initialize a null list
    unique_list = []
  
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    # print list
    for x in unique_list:
        print(x)
  
unique(df_tweets['created_at'])
```

```python
# Obtain timestamp in a readable format
from datetime import datetime
from datetime import timedelta
TIMESTAMP = (datetime.strptime(UNTIL_DATE, "%Y-%m-%d")-timedelta(days=1)).strftime('%Y%m%d_%H%M%S')

# Define working path and filename
PATH = os.getcwd()
FILENAME = PATH + '/' + TIMESTAMP + '.csv'

# Store dataframe in csv with creation date timestamp
df_tweets.to_csv(FILENAME, index = False)
    
p_end = time.time()

print('Scraping has completed!')
print('Total time taken to scrap is {} minutes.'.format(round(p_end-p_start)/60, 2))
```
